```python
#Testando a interface 1
#Os cálculos estão no final do caderno MNUM2017

x=0,10,50,100,200,300,400,500,600
x=np.array(x)
#y=np.array(0)
print(x)

y=math.tan(math.pi*0.0781)*x+619.57 #Corresponde ao coeficiente angular de 14,06°

print(y)

plt.plot(x,y,'k')
plt.xlabel("nx", fontsize=10)
plt.ylabel("nz", fontsize=10)
plt.grid(True)
```
